```python
df = pd.DataFrame(response.json().get('data').get('rows'), columns = response.json().get('data').get('header')).reset_index(drop=True)
df = df.iloc[:-1]
df
```

```python
df.info()
```

```python
columnas_a_mantener = [
    'Date',               # <-- porque de aquí generamos Date_num
    'Avg Temp ºC',
    'Avg DEW PT ºC',
    'Avg Wind Speed km/h',
    'Max wind Speed km/h',
    'Pressure HPA',
    'Precip. mm',
    'ET mm',
    'Wind dir'
]
```

```python
# Primero dejamos solo las columnas útiles
df = df[columnas_a_mantener]
```

```python
df.info()
```

```python
df['Date_num'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce').apply(lambda x: x.timestamp())
```

```python
df = df.drop(columns=['Date'])
df.info()
```

```python
df = df.rename(columns={
        'Avg Temp ºC': 'TempOut',
        'Avg DEW PT ºC': 'DewPt_',
        'Avg Wind Speed km/h': 'WSpeed',
        'Max wind Speed km/h': 'WHSpeed',
        'Pressure HPA': 'Bar',
        'Precip. mm': 'Rain',
        'ET mm': 'ET',
        'Wind dir': 'WDir_deg'
    })
```
